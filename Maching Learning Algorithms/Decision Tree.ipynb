```python
cirrohosis = pd.read_csv("C:/Users/minjk/Downloads/cirrhosis+patient+survival+prediction+dataset-1/cirrhosis.csv")
data = pd.DataFrame(cirrohosis)
#data.Drug.value_counts(dropna = False) 
# make pie chart
```

```python
# Question: We try to decide whether a primary biliary cirrhosis patient will die or survive (censored) from the previous dataset from 312 patients 
# data munging: To make a decision tree, all data has to be numerical
subset = data[["Status","Age", "Drug", "Stage"]]
subset = subset[subset["Drug"].notna()]
for i in range(0,312): 
    if subset.Status[i] == 'D': # 0: Death
       subset.Status[i] = 0
    elif subset.Status[i] == 'C': # 1: Censored
        subset.Status[i] = 1
    elif subset.Status[i] == 'CL': # 2: Censored due to liver transplantation
        subset.Status[i] = 2
subset.head()
```
