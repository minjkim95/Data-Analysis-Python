#### This notebook is for practice in K-Means Clustering

```python
import matplotlib.pyplot as plt
```

```python
x = [4,5,10,4,3,11,14,6,10,12]
y = [21,19,24,17,16,25,24,22,21,21]

plt.scatter(x,y)
plt.show()
```

```python
# Utilize Elbow method to visualize the inertia fro different values of K:
from sklearn.cluster import KMeans

data = list(zip(x,y))
inertias = []

for i in range(1,11):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(data) # train K-means model 
    inertias.append(kmeans.inertia_)

plt.plot(range(1,11), inertias, marker= 'o')
plt.title("Elbow method")
plt.xlabel("Number of Cluster")
plt.ylabel("Inertia") 
plt.show()

# THe elbow method shows that 2 is the elbow point which indicates a good value for k.
```

```python
kmeans = KMeans(n_clusters= 2) # Perform k-means clustering with 2 clusters
kmeans.fit(data) # train k-means model 

plt.scatter(x,y, c = kmeans.labels_)
plt.show() 
```

```python
kmeans = KMeans(n_clusters= 3) # Perform k-means clustering with 3 clusters
kmeans.fit(data) # train k-means model 

plt.scatter(x,y, c = kmeans.labels_)
plt.show()
```
